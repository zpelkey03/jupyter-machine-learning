```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

jump_statistics = pd.read_excel(r'\data_for_CURE.xlsx')
```

```python
jump_statistics = jump_statistics.rename(columns = { jump_statistics.columns[3]: "person_jumping"})
```

```python
sorted1 = jump_statistics.sort_values('person_jumping', ignore_index = True);
```

```python
f0,f1,f2 = sorted1.person_jumping.value_counts()
f0,f1,f2
```

```python
fig, axs = plt.subplots(   figsize =(13, 2.5))

axs1 = plt.subplot2grid(  shape= (1,4), loc = (0,0))
axs2 = plt.subplot2grid(  shape= (1,4), loc = (0,1))
axs3 = plt.subplot2grid(  shape= (1,4), loc = (0,2))

plt.tight_layout()

axs1.hist(sorted1.iloc[0:f0,0], edgecolor='b', fc='none', label = 'Zach Pelkey')
axs1.hist(sorted1.iloc[f0:f0+f1,0], edgecolor='r', fc='none', label = 'Jacob Payne')
axs1.hist(sorted1.iloc[f0+f1:f0+f1+f2,0], edgecolor='g', fc='none', label = 'Jacob Pull')
axs1.set_xlabel ('jump height')
axs1.set_ylabel ('Frequency')
axs1.legend()

axs2.scatter (sorted1.iloc[0:f0,1], sorted1.iloc[0:f0,0], label = 'Zach Pelkey')
axs2.scatter (sorted1.iloc[f0:f0+f1,1], sorted1.iloc[f0:f0+f1,0], label = 'Jacob Payne')
axs2.scatter (sorted1.iloc[f0+f1:f0+f1+f2,1], sorted1.iloc [f0+f1:f0+f1+f2,0], label = 'Jacob Pull')
axs2.set_xlabel ('jump distance')
axs2.set_ylabel ('jump height')
axs2.legend()

axs3.scatter (sorted1.iloc[0:f0,2], sorted1.iloc[0:f0,0], label = 'Zach Pelkey')
axs3.scatter (sorted1.iloc[f0:f0+f1,2], sorted1.iloc[f0:f0+f1,0], label = 'Jacob Payne')
axs3.scatter (sorted1.iloc[f0+f1:f0+f1+f2,2], sorted1.iloc[f0+f1:f0+f1+f2,0], label = 'Jacob Pull')
axs3.set_xlabel ('backwards jump distance')
axs3.set_ylabel ('jump height')
axs3.legend()
```
